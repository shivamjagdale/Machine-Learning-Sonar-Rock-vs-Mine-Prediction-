```python
# libraries

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
# importing our datasets

a = pd.read_csv('/content/drive/MyDrive/ML Datasets/sonar-data.csv', header=None)

sonar = pd.DataFrame(a)
```

```python
sonar   #checking the data
```

```python
# looking at the number of data we have

sonar[60].value_counts()
```

```python
# grouping the data by the target to see their dependencies

sonar.groupby(60).mean()
```

```python
# seperating the data based on labels and targets.

X = sonar.drop(columns=60)
```

```python
Y = sonar[60]
```

```python
# splitting the data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, stratify=Y, random_state=2)
```

```python
# taking a look at the data before and after the split

print(X.shape, X_train.shape, X_test.shape)
```

```python
# loading the Logistic Regression

model = LogisticRegression()
```

```python
# fitting the model
model.fit(X_train, Y_train)
```

```python
# predicting

X_train_prediction = model.predict(X_train)
```

```python
# checking the accuracy score

X_train_accuracyscore = accuracy_score(X_train_prediction, Y_train)
```

```python
print(X_train_accuracyscore)
```

```python
# same for testing data

X_test_prediction = model.predict(X_test)
```

```python
X_test_accuracyscore = accuracy_score(X_test_prediction, Y_test)
```

```python
print(X_test_accuracyscore)
```
